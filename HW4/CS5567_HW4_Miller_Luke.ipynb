## Importing Libraries

```python
import csv
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression


import warnings
warnings.filterwarnings("ignore")
```

## Data Importing

```python
X = []
Y = []

with open('/content/data.txt','r') as f:
    reader = csv.reader(f)
    for row in reader:
        X.append(float(row[0]))
        Y.append(float(row[1]))
```

## Plotting Data

```python
# plot parameter setting
fig,ax= plt.subplots(figsize =(8,8))
fig.patch.set_facecolor('#f5f6f6')

ax.set_facecolor('#f5f6f6')
ax.axes.get_yaxis().set_visible(True)
ax.axes.get_xaxis().set_visible(True)

for loc in ['right', 'top']:
    ax.spines[loc].set_visible(False)


scatter_size = [val * 15 for val in Y]

# plotting the values
ax.scatter(X,Y,s =scatter_size ,c= '#C2BB00',edgecolor ='black')


# labeling and titles
ax.set_title('Python Demo: Linear Regression',{'font':'serif', 'size':14, 'weight':'bold', })
ax.set_xlabel('X values',{'font':'serif', 'size':12, 'weight':'bold', })
ax.set_ylabel('Y values',{'font':'serif', 'size':12, 'weight':'bold', })



fig.show()
```

## Linear Regression Model Training

```python
#  splitting data into train,testing

X = np.array(X).reshape(-1,1)
Y = np.array(Y).reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25)

# Model instance
regr = LinearRegression()

# training the model
regr.fit(X_train, y_train)


# metrics and scores
print(regr.score(X_test, y_test))
```

```python
# predictions

y_preds = regr.predict(X_test)
```

### Plotting the results

```python
# plot parameter setting
fig,ax= plt.subplots(figsize =(10,10))
fig.patch.set_facecolor('#f5f6f6')

ax.set_facecolor('#f5f6f6')
ax.axes.get_yaxis().set_visible(True)
ax.axes.get_xaxis().set_visible(True)

for loc in ['right', 'top']:
    ax.spines[loc].set_visible(False)


scatter_size_true = [(val+1) * 15 for val in y_test]
scatter_size_preds = [(val+1) * 15 for val in y_preds]

# plotting the values
ax.scatter(X_test,y_test,s =scatter_size_true ,c= '#C2BB00',edgecolor ='black')
ax.scatter(X_test, y_preds,s =scatter_size_preds ,c= '#E1523D',edgecolor ='black')
ax.plot(X_test,y_preds,**{'linewidth':0.5,'linestyle': '--', 'color' : '#003547'}, alpha = 0.8)

# labeling and titles
ax.set_title('Python Demo: Linear Regression',{'font':'serif', 'size':14, 'weight':'bold', })
ax.set_xlabel('X values',{'font':'serif', 'size':12, 'weight':'bold', })
ax.set_ylabel('Y values',{'font':'serif', 'size':12, 'weight':'bold', })

# label
ax.text(10.5, 0,'Ground Truth',{'font':'serif', 'size':13, 'color':'#C2BB00', 'weight':'bold'}, alpha = 1)
ax.text(12.35,0 , '|',{'font':'serif', 'size':13, 'color':'grey','weight':'bold'}, alpha = 1)
ax.text(12.5, 0,'Prediction',{'font':'serif', 'size':13, 'color':'#E1523D', 'weight':'bold'}, alpha = 1)


fig.show()
```
